```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("sales_data.csv")

print("========== SALES DATA ==========")
print(df)

# -------------------------------
# Basic Statistics
# -------------------------------

print("\nTotal Sales:", df["Sales"].sum())
print("Average Sales:", df["Sales"].mean())
print("Highest Sales:", df["Sales"].max())
print("Lowest Sales:", df["Sales"].min())

# Highest sales day
highest = df.loc[df["Sales"].idxmax()]
print("\nHighest Sales Day")
print(highest)

# Lowest sales day
lowest = df.loc[df["Sales"].idxmin()]
print("\nLowest Sales Day")
print(lowest)

# Average Advertisement Spend
print("\nAverage Advertisement Spend")
print(df[["TV_Ads","Social_Media_Ads","Newspaper_Ads"]].mean())

# Total Advertisement Spend
df["Total_Ads"] = df["TV_Ads"] + df["Social_Media_Ads"] + df["Newspaper_Ads"]

print("\nAdvertisement Spend Per Day")
print(df[["Day","Total_Ads"]])

# Days with sales greater than 1500
print("\nSales Greater Than 1500")
print(df[df["Sales"] > 1500])

# Correlation
print("\nCorrelation Matrix")
print(df[["TV_Ads","Social_Media_Ads","Newspaper_Ads","Sales"]].corr())

# ----------------------------------
# Charts
# ----------------------------------

# Sales Trend
plt.figure(figsize=(8,5))
plt.plot(df["Day"], df["Sales"], marker='o')
plt.title("Daily Sales Trend")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.grid(True)
plt.show()

# Advertisement Comparison
plt.figure(figsize=(8,5))
plt.bar(df["Day"], df["TV_Ads"], label="TV Ads")
plt.bar(df["Day"], df["Social_Media_Ads"], bottom=df["TV_Ads"], label="Social Media")
plt.bar(df["Day"], df["Newspaper_Ads"], bottom=df["TV_Ads"]+df["Social_Media_Ads"], label="Newspaper")
plt.title("Advertisement Spend")
plt.xlabel("Day")
plt.ylabel("Amount")
plt.legend()
plt.show()

# Scatter Plot
plt.figure(figsize=(8,5))
plt.scatter(df["TV_Ads"], df["Sales"])
plt.title("TV Ads vs Sales")
plt.xlabel("TV Advertisement")
plt.ylabel("Sales")
plt.grid(True)
plt.show()

# Pie Chart
totals = [
    df["TV_Ads"].sum(),
    df["Social_Media_Ads"].sum(),
    df["Newspaper_Ads"].sum()
]

labels = ["TV Ads","Social Media","Newspaper"]

plt.figure(figsize=(6,6))
plt.pie(totals, labels=labels, autopct="%1.1f%%")
plt.title("Advertisement Distribution")
plt.show()

# ----------------------------------
# Business Insights
# ----------------------------------

print("\n========== BUSINESS INSIGHTS ==========")

print("1. Total Revenue Generated:", df["Sales"].sum())

print("2. Best Sales Day:", int(highest["Day"]))

print("3. Highest Sales:", highest["Sales"])

print("4. Lowest Sales Day:", int(lowest["Day"]))

print("5. Average Daily Sales:", round(df["Sales"].mean(),2))

print("6. Average Total Advertisement Spend:", round(df["Total_Ads"].mean(),2))

if df["TV_Ads"].corr(df["Sales"]) > 0.8:
    print("7. TV Advertisement has a strong positive relationship with Sales.")

print("\nProject Completed Successfully!")
```
